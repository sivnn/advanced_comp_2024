# Data Engineering

## Merge Tables

```python
# merge for breed 1
df_train = df_train.merge(df_breeds[['BreedID','BreedName']], how ='left' ,left_on='Breed1', right_on='BreedID').drop('BreedID',axis=1).rename(columns= {'BreedName':'Breed_1_Name'})

# merge for breed 2
df_train = df_train.merge(df_breeds[['BreedID','BreedName']], how ='left', left_on='Breed2', right_on='BreedID').drop('BreedID',axis=1).rename(columns= {'BreedName':'Breed_2_Name'})

# merge for color 1
df_train = df_train.merge(df_colors, how ='left', left_on='Color1', right_on='ColorID').drop('ColorID',axis=1).rename(columns= {'ColorName':'color_1_Name'})

# merge for color 2
df_train = df_train.merge(df_colors, how ='left', left_on='Color2', right_on='ColorID').drop('ColorID',axis=1).rename(columns= {'ColorName':'color_2_Name'})

# merge for color 3
df_train = df_train.merge(df_colors, how ='left', left_on='Color3', right_on='ColorID').drop('ColorID',axis=1).rename(columns= {'ColorName':'color_3_Name'})

# merge state
df_train = df_train.merge(df_states, how ='left', left_on='State', right_on='StateID').drop(['StateID','State'],axis=1)
```

## Data Cleaning

```python
# Gender - change to category
df_train['Gender'] = df_train['Gender'].replace({1:'Male',2:'Female',3:'Mixed'})

# Vaccinated - change 'not sure' (3) to not vaccinated (2) as when in doubt they will probably have to vaccinated again
df_train['Vaccinated'] = df_train['Vaccinated'].replace({3:2})

# Dewormed - same as vaccinated
df_train['Dewormed'] = df_train['Dewormed'].replace({3:2})

# Sterilized - while it is easy to figure out once you get your pet, not being sure is like assuming you pet has not been spayed\neutered
df_train['Sterilized'] = df_train['Sterilized'].replace({3:2})

# Remove breeds that have less than 20 samples
breeds = df_train['Breed_1_Name'].value_counts()
breeds_to_keep = breeds[breeds >= 20].index.to_list()

new_breeds_list = []
for x in df_train['Breed_1_Name']:
  if x in breeds_to_keep:
    new_breeds_list.append(x)

  else:
    new_breeds_list.append('other')

df_train['Breed_1_Name'] = new_breeds_list

df_train['Breed_1_Name'] = df_train['Breed_1_Name'].fillna('')
# df_train['Breed_2_Name'] = df_train['Breed_2_Name'].fillna('')
# df_train['color_2_Name'] = df_train['color_2_Name'].fillna('')
# df_train['color_3_Name'] = df_train['color_3_Name'].fillna('')
desc_ind = df_train['Description'][~(df_train['Description'].str.len() >= 4)].index
df_train['Description'][desc_ind] = ''

# there are few animals that were adopted on the same day (0)
df_train['AdoptionSpeed'] = df_train['AdoptionSpeed'].replace({0:1,3:2,4:3})

# states
df_train['StateName'] = df_train['StateName'].replace({'Johor':'other','Perak':'other','Negeri Sembilan':'other',
                                                           'Melaka':'other','Kedah':'other','Pahang':'other','Terengganu':'other',
                                                           'Sabah':'other','Kelantan':'other','Sarawak':'other','Labuan':'other'})

# drop columns that are not needed
columns_to_drop = ['Name','Breed1','Breed2','Color1','Color2','Color3','RescuerID','PetID']
df_train = df_train.drop(columns_to_drop, axis = 1)
```

### Sentiment Analysis

```python
#define function to detect and translate text if needed
def detect_and_translate(text,target_lang = 'en'):

  if text == '':
    return ''

  else:
    result_lang = detect(text)
    if result_lang == target_lang:
        return text

    else:
        translator = Translator()
        translate_text = translator.translate(text,  dest=target_lang)
        return translate_text.text
```

```python
#define function to detect and translate text if needed
def detect_lang(text,target_lang = 'en'):

  if text == '' or not text:
    return('empty')

  else:

    return detect(text)
```

```python
def feature_graph(df,feature,color1='darkgoldenrod',
                  color2='gold',color3='lightyellow',
                  sort_value = False, label=''):
  df_pivot = df[[feature,'AdoptionSpeed','Age']].pivot_table(index=feature,
                                                             columns = 'AdoptionSpeed',
                                                             values='Age',
                                                             aggfunc='count')
  df_pivot = df_pivot.div(df_pivot.sum(axis=1), axis=0)
  df_pivot = df_pivot.dropna(axis = 0 , how = 'any')

  if sort_value:
    df_pivot = df_pivot.sort_values(by=1)

  if label == '':
    la = f"Adoption Speed by Animal's {feature}"
  else:
    la = label


  # create data
  x = df_pivot.index
  y1 = df_pivot[1]
  y2 = df_pivot[2]
  y3 = df_pivot[3]


  # plot bars in stack manner
  plt.bar(x, y1, color=color1)
  plt.bar(x, y2, bottom=y1, color=color2)
  plt.bar(x, y3, bottom=y1+y2, color=color3)
  plt.xlabel(f'{feature}')
  plt.ylabel("Adoption Speed")
  plt.legend(["Up to a week", "Up to 3 months", "Over 100 days"])
  plt.title(la)
  plt.xticks(rotation = 90)
  plt.show()
```

### activation functions

```python
# batch size check
import timeit
start = timeit.default_timer()

activation_func = ['LeakyReLU','ReLU','Tanh','Sigmoid','Softmax']

accuracy_of_batch_activation = []

for a in activation_func:
  data_config = DataConfig(
      target= target,
      continuous_cols=con_cols,
      categorical_cols=cat_cols,
      continuous_feature_transform="quantile_normal",
      normalize_continuous_features=True
  )
  trainer_config = TrainerConfig(
      # auto_lr_find=True,  # Runs the LRFinder to automatically derive a learning rate
      batch_size=128,
      max_epochs=100,
  )
  optimizer_config = OptimizerConfig()

  model_config = CategoryEmbeddingModelConfig(
      # task="regression",
      task = "classification",
      layers="4096-4096-512",  # Number of nodes in each layer
      # activation="LeakyReLU",
      activation= a ,
      learning_rate=1e-3,
      metrics=["accuracy"],
  )

  tabular_model = TabularModel(
      data_config=data_config,
      model_config=model_config,
      optimizer_config=optimizer_config,
      trainer_config=trainer_config,
  )

  tabular_model.fit(train=train)
  result = tabular_model.evaluate(test)
  stop = timeit.default_timer()

  print()
  print(f'Finished batch {a}, Run time: {round((stop - start)/60,2)} Minutes')
  print()

  accuracy_of_batch_activation.append(result[0]['test_accuracy'])

```

## Embedding

https://www.kaggle.com/code/classtag/cat2vec-powerful-feature-for-categorical

```python
def apply_w2v(sentences, model, num_features):
    def _average_word_vectors(words, model, vocabulary, num_features):
        feature_vector = np.zeros((num_features,), dtype="float64")
        n_words = 0.
        for word in words:
            if word in vocabulary:
                n_words = n_words + 1.
                # feature_vector = np.add(feature_vector, model[word])
                feature_vector = np.add(feature_vector, model.wv[word])

        if n_words:
            feature_vector = np.divide(feature_vector, n_words)
        return feature_vector

    vocab = set(model.wv.index_to_key)
    feats = [_average_word_vectors(s, model, vocab, num_features) for s in sentences]
    return np.array(feats)
```

```python
def gen_cat2vec_sentences(data):
    X_w2v = copy.deepcopy(data)
    names = list(X_w2v.columns.values)
    for c in names:
      X_w2v[c]= [f'{c}_{g}' for g in X_w2v[c]]
      X_w2v[c] = X_w2v[c].fillna('unknow').astype('category')

    X_w2v = X_w2v.values.tolist()
    return X_w2v
```

```python
n_cat2vec_feature  = len(cat_cols) # define the cat2vecs dimentions
n_cat2vec_window   = len(cat_cols) * 2 # define the w2v window size
```

```python
def fit_cat2vec_model():
    X_w2v = gen_cat2vec_sentences(train[cat_cols].sample(frac=0.6))
    for i in X_w2v:
        shuffle(i)
    model = Word2Vec(sentences = X_w2v, vector_size=n_cat2vec_feature, window=n_cat2vec_window)
    return model

c2v_model = fit_cat2vec_model()
```

```python
train_c2v_matrix = apply_w2v(gen_cat2vec_sentences(train[cat_cols]), c2v_model, n_cat2vec_feature)
test_c2v_matrix = apply_w2v(gen_cat2vec_sentences(test[cat_cols]), c2v_model, n_cat2vec_feature)
```
